```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Load the tokenizer and model
model_name = "w11wo/indonesian-roberta-base-sentiment-classifier"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create a pipeline
sentiment = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Perform sentiment analysis and store results in the DataFrame
results = []
for text in df['full_text']:
    result = sentiment(text)[0]  # Get the first element from the pipeline output
    results.append(result['label']) # Append only the sentiment label to the results list

df['sentiment'] = results #add a new column in the dataframe called sentiment and assign the values from the result list.

# Pie Chart: Distribusi Sentimen
def plot_pie_chart(df):
    sentiment_counts = df['sentiment'].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
    plt.title('Distribusi Sentimen (Pie Chart)')
    plt.axis('equal')
    plt.show()

# Countplot / Barplot: Jumlah per Sentimen
def plot_bar_chart(df):
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='sentiment', palette='pastel')
    plt.title('Distribusi Sentimen (Barplot)')
    plt.xlabel('Sentimen')
    plt.ylabel('Jumlah Teks')
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# WordCloud per Sentimen
def plot_wordcloud_by_sentiment(df, sentiment_label):
    # Filter the DataFrame for the specified sentiment label
    filtered_df = df[df['sentiment'] == sentiment_label]

    # Check if the filtered DataFrame is empty
    if filtered_df.empty:
        print(f"No tweets found with sentiment: {sentiment_label}")
        return  # Exit the function if no tweets are found

    text = ' '.join(filtered_df['full_text'].astype(str))
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='Set2').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'WordCloud untuk Sentimen: {sentiment_label}')
    plt.show()

# 🔧 Panggil fungsi-fungsi berikut sesuai kebutuhan
plot_pie_chart(df)
plot_bar_chart(df)

# Untuk wordcloud berdasarkan label (positive/neutral/negative)
plot_wordcloud_by_sentiment(df, 'positive')
plot_wordcloud_by_sentiment(df, 'neutral')
plot_wordcloud_by_sentiment(df, 'negative')
```
